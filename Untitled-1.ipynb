```python
X4_ = caption4.reshape(-1, 1)
f"Tipo: {type(X4_)} - Dimensiones: {X4_.ndim} - Longitud: {len(X4_)} - Longitud dim 0: {len(X4_[0])} - Longitud dim 1: {len(X4_[1])}"
```

```python
dbscan4 = DBSCAN(eps=0.002, min_samples=50, metric='euclidean', n_jobs=-1)
dbscan4.fit(X4_)
unique_labels = np.unique(dbscan4.labels_)
unique_labels
```

```python
# Obtener las coordenadas de los pixeles que pertenecen a cada cluster
for label in unique_labels:
    if label == -1:
        # Ignorar los puntos ruidosos
        continue
    # Crear una m√°scara booleana para los puntos que pertenecen al cluster actual
    mask = dbscan4.labels_ == label
    # Obtener las coordenadas de los puntos que pertenecen al cluster actual
    coords = np.argwhere(mask)
    # Resaltar los puntos en la imagen original
    caption4[coords[:, 0], coords[:, 1]] = [255, 0, 0]  # Rojo brillante
```
